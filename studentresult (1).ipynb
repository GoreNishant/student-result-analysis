# visualization and analyzing

```python
colors=['pink','blue']
ax=sns.countplot(data=df,x='Gender',palette=colors)
plt.title("gender wise count of students")

for container in ax.containers:
    ax.bar_label(container)
```

as you seee in above figure female students more as compare to male

```python
gb=df.groupby('ParentEduc')[['MathScore','ReadingScore','WritingScore']].agg({'MathScore':np.mean,'ReadingScore':np.mean,'WritingScore':np.mean})
```

# analyzing imapact of parent education on student scores

```python
sns.heatmap(gb,annot=True)
plt.title("parent education wise students average mark")
```

as you see in above figure that master degree educated parents children get good marks average.

so we anlyze from above figure that parents education impact on students avg score

```python
df_long = pd.melt(df, id_vars=['Gender'], value_vars=['MathScore', 'ReadingScore', 'WritingScore'],
                  var_name='Subject', value_name='Score')


plt.figure(figsize=(8, 6))
sns.boxplot(x='Subject', y='Score', hue='Gender', data=df_long)


plt.title('Gender-wise Score Distribution')
plt.xlabel('Subject')
plt.ylabel('Score')


plt.show()
```

#as you see in above figure boys are little good in math than girls but girls are good in read and write
